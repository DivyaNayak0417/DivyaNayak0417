```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']= (12,8) #Adjusts the configuration of the plots 
```

```python
# Read the data
df = pd.read_csv('movies.csv')
# look at the data
df.head()
```

```python
# Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} -{}%'.format(col, pct_missing))
# remove 'na'

df = df.dropna()
```

```python
# Data types of columns
df.dtypes
```

```python
# Chaning the data type of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create correct year column
df['year_correct']= df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
df['year_correct']
```

```python
df.sort_values(by=['gross'], inplace = True, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Check and drop duplicates
df.drop_duplicates().head()
```

```python
# Checking for correlation
# Scatterplot - budget vs gross
plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot budget versus gross earnings using seaborn
sns.regplot(x= 'budget', y= 'gross', data = df,scatter_kws = {"color": "red"}, line_kws = {"color": "black"})
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
correlation_matrix= df.corr() # check for correlation for numeric features
# High correlation between Budget and Gross
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at Campany - Non Numeric
df.head()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]=  df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix= df_numerized.corr() # check for correlation for numeric features
# High correlation between Budget and Gross
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs_sorted= corr_pairs.sort_values()

high_correlation = corr_pairs_sorted[(corr_pairs_sorted) >0.5]
high_correlation
```
